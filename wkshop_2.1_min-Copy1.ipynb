## Exercise 5: Debt Calculator

Write a Python program to compute the amount of the debt in n months. The borrowing amount is $100,000 and the loan adds 5% interest of the debt and rounds it to the nearest 1,000 above month by month.

The function takes in an integer (the number of months) and returns the total debt including the starting \$100,000 rounded to \$1000 .

```
debt_total(7) -> 144000
debt_total(8) -> 152000
debt_total(9) -> 160000
debt_total(1) -> 105000
debt_total(57) -> 1736000
```


````python
import math
def debt_total(n):
    PV = 100000
    x = 1
    while x <= n:
        PV = PV*1.05
        PV = int(math.ceil(PV/1000.0))*1000
        x += 1
    return PV
print(debt_total(7))
print(debt_total(8))
print(debt_total(9))
print(debt_total(1))
print(debt_total(57))
````

Fizzbuzz is a game for children and a drinking game for adults. It's also a classic programming interview question.

Suppose we have a number n. We have to display a string representation of all numbers from 1 to n, but there are some constraints.

    If the number is divisible by 3, write Fizz instead of the number
    If the number is divisible by 5, write Buzz instead of the number
    If the number is divisible by 3 and 5 both, write FizzBuzz instead of the number

So the results are `1, 2, Fizz, 4, Buzz, ..., 14, Fizzbuzz, 16, 17, Fizz, 19, Buzz, ...`

Write a python program that prints fizzbuzz for numbers 1 to 150

````python
# exercise 7
def fizzbuzz (x):
    x = range(1,151)
    FB = []
    for item in x:
        if item %3 == 0 and item %5 == 0:
            FB.append("FizzBuzz")
        elif item %5 == 0:
            FB.append("Buzz")
        elif item %3 == 0:
            FB.append("Fizz")
        else:
            FB.append(item)
    return(FB)
fizzbuzz(x)
````

## Exercise 8: Palindrome?

A palindrome is a word which reads the same backward as forwards. Write a function that returns True if a string is a palindrome:

    “laval” ⇒ Yes

````python
# O(log(n))

def binary_search(array,element):
    """
    Return the index of a given number within an array
    
    :param[array,element]: it checks if element = mid value(in length of array), if larger than cuts first half of list.
    repeats the process, keeps cutting array in half till left with element and its index.
    :type[array, element]:array, element
    
    ...
    :return: index of element
    :rtype: int
    """

    mid = 0 
    start = 0
    end = len(array)
    step = 0

    while (start <= end):
        step += 1
        mid = (start + end) // 2

        if element == array[mid]:
            return mid

        if element < array[mid]:
            end = mid - 1

        else:
            start = mid + 1

    return -1

element = 18

array = [1,2,5,7,13,15,16,18,20,28,29]

binary_search(array, element)
````

````python
# O(nlog(n))

def mergeSort(X):
  '''
  sorting lists 
  
  :param[X]: the function goes through the list, sorts from the left and to the right,merges the results and moves 
  forward in the list repeating the process. after some accumulation, it goes back to the begining, sorts the build up
  (using the same process, going through all the objects). Does that till it gets to the end. 
  the final step is checking if any objects were missed in the merge, by verifying length of both lists are equivalent. 
  
  :type[X]:container of iterable objects. #the lecture example specifies a list of numbers, but the code would work 
                                          #with lists of strings or booleans. 
  ...
  :raises [ErrorType]: argument needs to be a container with >1 object. 
  ...
  :return: sorted argument
  :rtype: same container as input. 
  '''

  if len(X) > 1:
    mid = len(X)//2
 
    L = X[:mid]
    R = X[mid:]
    
    #Sort L
    mergeSort(L)
    #Sort R
    mergeSort(R)
    
    i = j = k = 0
    
    # add data to temp arrays L and R
    while i < len(L) and j < len(R):
        
        if L[i] < R[j]:
            X[k] = L[i]
            i += 1
        else:
            X[k] = R[j]
            j += 1
        k += 1
    
    #Last check to see if any items were left
    while i < len(L):
        X[k] = L[i]
        i += 1
        k += 1
    while j < len(R):
        X[k] = R[j]
        j += 1
        k += 1
    return(X)

num = [1,5,2,7,3,8]
mergeSort(num)
````

````python
# O(2^n)

def fibonacci(n):
    '''
    finds fibonacci equivalent to n. n considered as index within fibonacci sequence. 
  
    :param[n]: the code finds the two previous entries ((n-1)&(n-2)) and their fib equivalent, finally their sum.
    sum = fib equivalent of n
    :type[n]: int
    ...
    :raises [ErrorType]: argument has to be positive. 
    ...
    :return: fibonacci equivalent of n
    :rtype: int  
    '''
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

fibonacci(10)
````
